```python
import json
import requests
from config import nps_api_key
import pandas as pd
import hvplot.pandas
```

```python
url = 'https://developer.nps.gov/api/v1/'
category = 'feespasses'
park = 'yose'

query_url = url + category + '?parkCode=' + park + '&api_key=' + nps_api_key
```

```python
nps_response = requests.get(query_url)
nps_json = nps_response.json()

# nps_json
```

```python
fee_passes = nps_json['data'][0]['fees']
# fee_passes
```

```python
entrance = fee_passes[0]['entranceFeeType']
entrance
```

```python
cost = fee_passes[0]['cost']
cost
```

```python
len(fee_passes)
```

```python
entrances = []
costs = []

for i in range(10):
    entrances.append(fee_passes[i]['entranceFeeType'])
    costs.append(fee_passes[i]['cost'])
```

```python
entrances
```

```python
costs
```

```python
yose_feepass = pd.DataFrame({'Entrance': entrances,
                           'Cost': costs})
yose_feepass
```

```python
query2_url = url + 'parks?api_key=' + nps_api_key
```

```python
parks_response = requests.get(query2_url)
nps_parks_json = parks_response.json()

# nps_parks_json
```

```python
nparks = nps_parks_json['data']
name = nparks[0]['fullName']
name
```

```python
len(nparks)
```

```python
national_parks = []
lats = []
lons = []

for i in range(49):
    national_parks.append(nparks[i]['fullName'])
    lats.append(nparks[i]['latitude'])
    lons.append(nparks[i]['longitude'])
```

```python
df_nationalparks = pd.DataFrame({'National Park': national_parks,
                                 'Latitude': lats,
                                 'Longitude': lons})
```

```python
query_np_url = url + "parks?limit=1000&q=National%20Park&api_key=" + nps_api_key
```

```python
np1000_response = requests.get(query_np_url)
np1000_json = np1000_response.json()

# np1000_json
```

```python
n1000parks = np1000_json['data']
fullname = n1000parks[0]['fullName']
fullname
```

```python
len(n1000parks)
```

```python
nparks2 = []
lats2 = []
lons2 = []

for i in range(452):
    nparks2.append(n1000parks[i]['fullName'])
    lats2.append(n1000parks[i]['latitude'])
    lons2.append(n1000parks[i]['longitude'])
```

```python
activities = []
for i in range(452):
    activities.append(len(np1000_json['data'][i]['activities']))
```

```python
park_activities = pd.DataFrame({'National Park': nparks2,
                              'Activity Amount': activities})
park_activities
```

```python
msk2 = park_activities['National Park'].str.contains('National Park')
```

```python
np_activities = park_activities.loc[msk2]
np_activities
```

```python
full_npark_df = np_location.merge(np_activities, how='outer', on='National Park')
full_npark_df
```

```python
full_npark_df.info()
```

```python
nationalparks_df = full_npark_df.astype({"Latitude": float, "Longitude": float, "Activity Amount":float})
```

```python
nationalparks_df.info()
```

```python
npark_hvplot = nationalparks_df.hvplot.points('Longitude',
                                           'Latitude',
                                           geo=True,
                                           tiles='OSM',
                                           color='National Park',
                                           scale=1,
                                           alpha=0.5,
                                           frame_width=700,
                                           frame_height=500)
npark_hvplot
```

```python
nationalparks_df.to_csv(r'Resources/np_location.csv')
```
